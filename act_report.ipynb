### Dog Stages and their Ratings

We have four dog stages; `pupper` which is a young dog who is yet to reach the `puppo` stage. They are still too immature for the weight if being a doggo. The next stage is a `puppo` basically the teenage years of a dogs stages. This is where you have a mix of childish antics of a pupper or the responsible actions of a `doggo`. The third stage being the `doggo`., the resonable and reliable best friend of man, this dog stage shows that getting a little older can still be adorable. We also have `no dog stage` since some of the text  didn't contain the any mention of the dog stages.

`floofer` is a dog with extreme lengths of hair so must be delivered his own stage. With all of these different stages, finding out the population of the dogs and what stages they fall in. As well as, identfying which tweets were more popular through the dog rating.

#### No Stage

Let's start with the dogs without any dog stages, we have 9 ratings for them. 0/10, 1/10, 2/10, 11/10, 12/10, 13/10, 14/10 & 20/20

```python
no_sta = report.query('dog_stage == "no stage"')
```

```python
no_sta['rating_numerator'].value_counts()
```

```python
no_stage = no_sta['rating_numerator'].value_counts()
```

```python
# Labels of pie sections
label = ['10','11','12','13','14', '2', '1', '0', '20']
 
 # Creating color parameters
colors = ( "Red", "Green", "Grey", "Blue", "Magenta", "Chartreuse", "Cyan", "White", "Burlywood")
 
# Wedge attributes
wp = { 'linewidth' : 1, 'edgecolor' : "black" }
explode = (0, 0, 0, 0, 0.4, 1.0, 1.8, 2.8, 3.8)
 
# Displaying the percentage
def func(pct, allvalues):
    absolute = int(pct / 100.*np.sum(allvalues))
    return "{:.1f}%".format(pct, absolute)
 
# Creating Appointments Attended piechart
fig, ax = plt.subplots(figsize =(12, 12))
wedges, texts, autotexts = ax.pie(no_stage,
                                  explode=explode,
                                  autopct = lambda pct: func(pct, no_stage),
                                  labels = label,
                                  shadow = True,
                                  colors = colors,
                                  startangle = 90,
                                  wedgeprops = wp,
                                  textprops = dict(color ="Black"))
 
plt.setp(autotexts, size = 10, weight ="bold")
ax.set_title("Rating for no dog stage")
```

As we can see that the tweets without any dog stages specificed have quite a equal distribution between 10, 11 & 12/10 ratings. With 10/10 being the most present rating in the tweets, a percentage of 28.7% of all tweets. 13/10 wasn't too far behind, with a 33.74% decrease between the 13/10 rating and the 12/10 rating.

With the higher and lower ratings (14/10, 20/10, 2/10, 1/10, 0/10); there is a collective percentage of 2.8% which shows that its very rare for a rating of 14, 20, 2, 1, 0 amongst the tweets present. It's quite tough to be cute enough to be rated higher than 13/10 or not appealing enough to be rated lower than 10/10. Thankfully, meaning that the average ratings that these dogs get are between 10/10, 11/10 and 12/10.

### Breed & Dog ratings

Here we will be looking at the coorelation between the dog breeds and the dog ratings. TO see which dog breeds had the highest ratings

```python
report.groupby('rating_numerator')['breed'].value_counts()
```

```python
breed_rate = report.groupby('rating_numerator')['breed'].value_counts()
```

```python
breed_rating = breed_rate.unstack(level='rating_numerator')
```

```python
breed_rating = breed_rating.fillna(0)
```

```python
breed_rating = breed_rating.astype(int)
```

```python
breed_rating
```

Now we can see the number of ratngs for each dog breed and make a observation as to which breed has what rating and those with high or low numbers within that rating.
